```python
from src.alien import alien
import mlrose_hiive
from mlrose_hiive import NNGSRunner,NNClassifier

from sklearn.metrics import accuracy_score
```

```python
obj=alien()
X_test, y_test,X_train, y_train=obj.dataset(modify_feature=True)
```

```python
grid_search_parameters = {
    'max_iters': [1000],                          # nn params
    'learning_rate': [1e-2],                       # nn params
    'activation': [mlrose_hiive.relu],            # nn params
    'restarts': [1],                             # rhc params
}

nnr = NNGSRunner(
    x_train=X_train,
    y_train=y_train,
    x_test=X_test,
    y_test=y_test,
    experiment_name='nn_test_rhc',
    algorithm=mlrose_hiive.algorithms.rhc.random_hill_climb,
    # algorithm=mlrose_hiive.algorithms.gd.gradient_descent,
    grid_search_parameters=grid_search_parameters,
    iteration_list=[1, 10, 50, 100, 250, 500, 1000],
    hidden_layer_sizes=[[10,10]],
    bias=True,
    early_stopping=True,
    clip_max=10,
    max_attempts=100,
    n_jobs=5,
    seed=123456,
    output_directory=None
)
run_stats_df, curves_df, cv_results_df, grid_search_cv = nnr.run()

y_test_pred = grid_search_cv.predict(X_test)
y_test_accuracy = accuracy_score(y_test, y_test_pred)
print(y_test_accuracy)
```
